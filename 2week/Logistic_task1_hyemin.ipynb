```python
data = pd.read_csv('sampled_data.csv') 
```

```python
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.iloc[:, :-1], data.iloc[:, -1], test_size=0.3, random_state=0)
```

```python
scaler = StandardScaler() # Scaling
scaler.fit_transform(data)
```

```python
data_new = pd.DataFrame(scaler.fit_transform(data), columns = ['time','V1','V2','V3','V4','V5','V6','V7','V8','V9','V10','V11','V12','V13','V14','V15','V16','V17','V18','V19','V20','V21','V22','V23','V24','V25','V26','V27','V28','Amount','class'])
```

```python
data_new
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test) # class 예측
```

```python
y_pred
```

```python
X_test_proba=classifier.predict_proba(X_test)
```

```python
classifier.score(X_test, y_test) # return mean accuracy
```

# 2- Confusion Matrix 찍어보기 → Mean Accuracy와 비교

```python
from sklearn.metrics import *
```

```python
f1_score(y_pred, y_test) # f1 score

#f1_score가 더 작은 확률을 가지고 있다.
```

```python
confusion_matrix(y_pred, y_test)
```

# 3- curoff 조정하기

```python
for thres in [0.5, 0.4, 0.3, 0.2, 0.1, 0.01]:
    print('cut-off: ', thres)
    print(confusion_matrix(y_test, X_test_proba[:,1]>thres))
    print('auc : ', roc_auc_score(y_test, X_test_proba[:,1]>thres))
    print('recall', recall_score(y_test, X_test_proba[:,1]>thres))
    print('-'*35)
    
#thres에 값을 넣어서 curoff를 조정했다.
```
