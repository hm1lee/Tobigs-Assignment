## 1. 예매순 상위 5개의 현재 상영 중인 영화 가져오기

영화 제목, 주연배우 3인

```python
def movie_title_url_actor():
    url = "https://movie.naver.com/movie/running/current.nhn"
    res = requests.get(url)
    html = res.text
    soup = BeautifulSoup(html, 'html.parser')
    
    #제목 추출
    titles = soup.select('dt[class="tit"] a', limit =5)
    title = []
    for k in titles:
        title.append(k.get_text())
    print(title)
    
    #배우 추출
    actors = []
    for actor in range(2, 15, 3):
        actors.append(soup.select('span[class="link_txt"]')[actor].select('a',limit=3))
    print(actors)
```

## 2. 해당 영화의 평점 가져오기

네티즌 평점, 관람객 평점, 기자/평론가 평점

```python
import re
```

```python
url=["https://movie.naver.com/movie/bi/mi/basic.nhn?code=179181","https://movie.naver.com/movie/bi/mi/basic.nhn?code=186821","https://movie.naver.com/movie/bi/mi/basic.nhn?code=187321","https://movie.naver.com/movie/bi/mi/basic.nhn?code=186613","https://movie.naver.com/movie/bi/mi/basic.nhn?code=181925"]
```

```python
def get_grade():
    url=["https://movie.naver.com/movie/bi/mi/basic.nhn?code=179181","https://movie.naver.com/movie/bi/mi/basic.nhn?code=186821","https://movie.naver.com/movie/bi/mi/basic.nhn?code=187321","https://movie.naver.com/movie/bi/mi/basic.nhn?code=186613","https://movie.naver.com/movie/bi/mi/basic.nhn?code=181925"]
    for i in url : 
        res = requests.get(i) 
        html = res.text
        soup = BeautifulSoup(html, 'html.parser')
        
        name = soup.find('h3', class_='h_movie')
        names = ''
        for i in name.find_all('a'):
            print(i.text)
        
        # 관람객 평점
        data = soup.find('div', class_='score') #div 태그의 score 클래스 추출
        data_a1 ='' 
        
        for i in data.find_all('em'): #data의 em태그 추출
            data_a1 += i.text #data_a에 저장
        
        data_a1 = data_a1[-4:]  
        p = re.compile('[0-9]{1}.[0-9]{2}') #정규표현식 이용
        k = p.findall(data_a1) #data_a에 p규칙 적용하기
        print('관람객 평점 :', k)
        
        # 네티즌 평점
        data = soup.find('div', class_='score score_left') #div 태그의 star_score 클래스 추출
        data_a2 ='' 
        
        for i in data.find_all('em'): #data의 em태그 추출
            data_a2 += i.text #data_a에 저장
        
        data_a2 = data_a2[-4:]  
        p = re.compile('[0-9]{1}.[0-9]{2}') #정규표현식 이용
        t = p.findall(data_a2) #data_a에 p규칙 적용하기
        print('네티즌 평점 :', t)
        
        # 기자/ 평론가 평점
        data = soup.find('div', class_='spc_score_area') #div 태그의 spc_score_area 클래스 추출
        data_a3 ='' 
        
        for i in data.find_all('em'): #data의 em태그 추출
            data_a3 += i.text #data_a에 저장
        
        data_a3 = data_a3[-4:]  
        p = re.compile('[0-9]{1}.[0-9]{2}') #정규표현식 이용
        s = p.findall(data_a3) #data_a에 p규칙 적용하기
        print('기자/평론가 평점 :', s)
        print('')
    
        
```
